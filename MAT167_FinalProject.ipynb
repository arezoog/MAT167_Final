### Build Raw Google Matrix

Creates a 12x12 matrix filled with 0.0 values. We then fill in non-zero entries one by one. G[i][j] = 1/k where k = number of outlinks from i. This distributes the 'vote weight' of page i equally across all pages it links to. A page with 4 outlinks contributes 0.25 to each; a page with 1 outlink contributes 1.0 to that single destination.
